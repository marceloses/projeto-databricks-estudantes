# Aplicando update na coluna Genero

```python
%sql
UPDATE estudantes.default.estudantes_silver
SET Genero = CASE
    WHEN Genero = 'Female' THEN 'Mulher'
    WHEN Genero = 'Male' THEN 'Homem'
    ELSE Genero
END;
```

# Selecionando a Tabela

```python
%sql
SELECT DISTINCT genero, nivel_escolaridade_pais, Nota_exame FROM estudantes.default.estudantes_silver
    
```

# Criando um Dashboard

```python
# Este código importa a biblioteca matplotlib para visualização de dados,
# lê a tabela 'estudantes_silver' do catálogo 'estudantes.default',
# agrupa o DataFrame pelo campo 'Genero' e conta o número de ocorrências de cada gênero.
# Em seguida, converte o resultado para um DataFrame pandas e cria um gráfico de pizza mostrando
# a distribuição percentual de cada gênero entre os estudantes.

import matplotlib.pyplot as plt

df_silver = spark.read.table("estudantes.default.estudantes_silver")
counts = df_silver.groupBy('Nivel_escolaridade_pais').count().toPandas()

plt.figure(figsize=(8,6))
plt.pie(
    counts['count'],
    labels=counts['Nivel_escolaridade_pais'],
    autopct='%.1f%%',
    startangle=90
)
plt.title("Distribuição de Gênero")
plt.axis('equal')
plt.show()
```
