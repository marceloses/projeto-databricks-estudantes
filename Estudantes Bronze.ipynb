# Criando tabela temporaria Camada Bronze

```python
# Criando uma tabela temporária do arquivo estudantes.
file_path = 'dbfs:/Volumes/estudantes/default/estudantes/Estudantes.csv'
df_estudantes = spark.read.format('csv') \
    .option('header', 'true') \
    .option('inferSchema', 'true')\
    .option('delimiter', ',') \
    .load(file_path)
df_estudantes.createOrReplaceTempView("temp_estudantes")

# Exibindo a tabela temporária
display(spark.sql("SELECT * FROM temp_estudantes"))
```

# Selecionando Tabela Temporária

```python
%sql
--SELECT * FROM temp_estudantes
SELECT genero, nivel_escolaridade_pais, nota_exame FROM temp_estudantes

```

# Filtrando Tabela Temporária 

```python
%sql
SELECT frequencia_escolar, Nivel_Motivacao, Renda_Familiar, Acesso_Internet, Genero,Nivel_escolaridade_pais,Nota_exame FROM temp_estudantes 
```

# Criando uma tabela fisica a partir da temporária

```python
%sql
-- Criando ou sobrescrevendo a tabela física "Camada Silver" a partir da tabela temporária "temp_estudantes" Camada Bronze
CREATE OR REPLACE TABLE estudantes.default.estudantes_silver AS
SELECT Genero, Nivel_escolaridade_pais, Nota_exame FROM temp_estudantes
```
